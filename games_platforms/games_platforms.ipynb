````python
platforms = df.groupby(by='platform')['total_sales'].\
agg('sum').sort_values(ascending=False).reset_index()
platforms
````

````python
# –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü part_sales, –≤ –Ω–µ–≥–æ –∑–∞–ø–∏—à–µ–º –¥–æ–ª—é –ø—Ä–æ–¥–∞–∂ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤ %

platforms['part_sales'] = platforms['total_sales'] / platforms['total_sales'].sum()
platforms['part_sales'] = platforms['part_sales'].round(3) * 100
````

ùêª1 - –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ—Ç–ª–∏—á–∞—é—Ç—Å—è.

–í–æ–∑—å–º–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº=5%

````python
# –∑–∞–¥–∞–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

alpha = 0.05 
````

````python
actual_df.platform.unique()
````

````python
# –¥–µ–ª–∞–µ–º —Å—Ä–µ–∑, –±–µ—Ä–µ–º —Ç–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏–π, =XOne –∏ !=np.NaN

xone_scores = actual_df.loc[(actual_df['platform'] == 'XOne') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
xone_scores.head()
````

````python
# —Ç–æ –∂–µ –¥–ª—è PC

pc_scores = actual_df.loc[(actual_df['platform'] == 'PC') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
pc_scores.head()
````

````python
# —Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫

display(f"–†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è XOne {len(xone_scores)}")
display(f"–†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è PC {len(pc_scores)}")
````

````python
# —Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é. –ú—ã —Ä–∞–±–æ—Ç–∞–µ–º —Å –≤—ã–±–æ—Ä–∫–∞–º–∏, ddof —Å—Ç–∞–≤–∏–º = 1

print('–î–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã Xone:', xone_scores.var(ddof=1))
print('–î–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã PC:', pc_scores.var(ddof=1))
````

````python
# –° –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ ttest_ind –≤—ã—á–∏—Å–ª—è–µ–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ 
# –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –µ–≥–æ —Å —É—Ä–æ–≤–Ω–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

results = st.stats.ttest_ind (xone_scores, pc_scores,equal_var=False)
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
````
