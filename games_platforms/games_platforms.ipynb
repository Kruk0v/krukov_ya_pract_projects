Сгруппируем данные по платформам и посчитаем продажи во всем мире

Данные подтверждют закон Паретто, 9 платформ дают 75% продаж. Сгруппированные данные запишем в переменную `platforms`, сделаем _reset_index_.

````python
platforms = df.groupby(by='platform')['total_sales'].\
agg('sum').sort_values(ascending=False).reset_index()
platforms
````

````python
# добавим столбец part_sales, в него запишем долю продаж платформы в %

platforms['part_sales'] = platforms['total_sales'] / platforms['total_sales'].sum()
platforms['part_sales'] = platforms['part_sales'].round(3) * 100
````

````python
# смотрим на таблицу с новым столбцом

platforms
````

````python
# какую долю занимают платформы с суммарными продажами < 1%:

platforms[platforms['part_sales'] < 1]['part_sales'].sum() 
````

Менее 1,5%. Удаляем эти платформы из `platforms`

````python
platforms = platforms[platforms.part_sales >= 1.0]
````

````python
platforms
````

Сохраним названия платформ с долей продаж >= 1% в список good_platforms

````python
good_platforms = list(platforms['platform'])
display(good_platforms)
display('Каждая из {:0} платформ составляют не менее 1% общих продаж'.format(len(good_platforms)))
````

В исходной таблице оставляем платформы с долей продаж > 1%.

````python
df = df.loc[df['platform'].isin(good_platforms)]
````

### 5. Проверка гипотез

#### 5.1 Проверка гипотезы о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем нулевую гипотезу:

𝐻0 - Средние пользовательские рейтинги платформ Xbox One и PC равны.

Сформулируем альтерантивную гипотезу:

𝐻1 - Средние пользовательские рейтинги платформ Xbox One и PC отличаются.

Возьмем уровень значимости 𝛼=5%

````python
# задаем уровень значимости

alpha = 0.05 
````

````python
actual_df.platform.unique()
````

````python
# делаем срез, берем только значений, =XOne и !=np.NaN

xone_scores = actual_df.loc[(actual_df['platform'] == 'XOne') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
xone_scores.head()
````

````python
# то же для PC

pc_scores = actual_df.loc[(actual_df['platform'] == 'PC') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
pc_scores.head()
````

````python
# смотрим на размеры выборок

display(f"Размер выборки пользовательских оценок для XOne {len(xone_scores)}")
display(f"Размер выборки пользовательских оценок для PC {len(pc_scores)}")
````

````python
# считаем дисперсию. Мы работаем с выборками, ddof ставим = 1

print('Дисперсия выборки пользовательских оценок для платформы Xone:', xone_scores.var(ddof=1))
print('Дисперсия выборки пользовательских оценок для платформы PC:', pc_scores.var(ddof=1))
````

````python
# С помощью функции ttest_ind вычисляем уровень значимости критерия Стьюдента 
# и сравниваем его с уровнем статистической значимости

results = st.stats.ttest_ind (xone_scores, pc_scores,equal_var=False)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

Нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC отвергнуть нельзя. Проверим средние оценки пользователей по XOne и PC

#### 5.2 Проверка гипотезы о том, что средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую гипотезу:

𝐻0 - Средние пользовательские рейтинги платформ Action и Sports равны.

Сформулируем альтерантивную гипотезу:

𝐻1 - Средние пользовательские рейтинги платформ Action и Sports отличаются.

Возьмем уровень значимости 𝛼=5%

````python
# задаем уровень значимости

alpha = 0.05 
````

Делаем по аналогии с предыдущей гипотезой

````python
actual_df.genre.unique()
````

````python
# делаем срез, берем только значений, =Action и !=np.NaN

action_scores = actual_df.loc[(actual_df['genre'] == 'Action') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
action_scores.head()
````

````python
# то же для Sports

sports_scores = actual_df.loc[(actual_df['genre'] == 'Sports') \
            & ~np.isnan(actual_df['user_score'])]['user_score']
sports_scores.head()
````

````python
# смотрим на размеры выборок

display(f"Размер выборки пользовательских оценок для Action {len(action_scores)}")
display(f"Размер выборки пользовательских оценок для Sports {len(sports_scores)}")
````

````python
# считаем дисперсию. Мы работаем с выборками, ddof ставим = 1

print('Дисперсия выборки пользовательских оценок для жанра Action:', action_scores.var(ddof=1))
print('Дисперсия выборки пользовательских оценок для жанра Sports:', sports_scores.var(ddof=1))
````

````python
# С помощью функции ttest_ind вычисляем уровень значимости критерия Стьюдента 
# и сравниваем его с уровнем статистической значимости

results = st.stats.ttest_ind (action_scores, sports_scores,equal_var=False)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
````

Нулевую гипотезу отвергаем, значит, средние пользовательские оценки для жанров Action и Sports статистически значимо различаются. Проверим, посчитаем средние оценки пользователей по таблице
