### Обработка пропусков

#### ceiling_height

```python
# Начнем с этого поля, в нем 9195 пропусков, проверим, какие значения в заполненных ячейках.

df['ceiling_height'].describe()
```

```python
# Аномально высокие потолки больше 3.0 метров:

df.query('ceiling_height > 3.0')['ceiling_height'].count()
```

```python
# Медиана и среднее значение очень близки, 3-й квартиль 2.8 м, значения в целом лежат в диапазоне до 3.0 м, 
# больше этого значения только у 1529 ячеек:
1529 / 23699 * 100
```

```python
df.query('ceiling_height > 3.0')['ceiling_height'].value_counts(normalize=True)
```

```python
# 6,5% цифра небольшая, количество больших значений тоже невелико, пропуски заменю медианой и проверю отсутсвие пропусков:

df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['ceiling_height'].isna().sum()
```

С пропусками в `ceiling_height` пока всё. Оставлять в столбце нули было бы неправильно. Когда в квартире ноль балконов, это не очень удобно, но не смертельно. Жить же в квартире с высотой потолков 0.0 м проблематично.
Пропуски возникли, скорее всего, из-за низкого приоритета этого параметра у подающих объявление. В приоритете местонахождение, количество комнат, площадь. Если потолок не очень высокий, декларировать это и не особо хочется.

#### floors_total

```python
df['floors_total'].describe()
```

Значения по столбцу тоже очень ровные, медиана и среднее близки, третий квартиль равен 16, пропусков всего лишь:

```python
86 / 23699 * 100
```

```python
# 0,36%, это очень мало. Удалять строчки не хочется, замена медианой при 0,36% от общего объема вряд ли исказит картину, 
# терять данные нежелательно, поэтому пропуски заменим медианой(медиана — целое число 9, никаких дополнительных действий
# делать не нужно):

df['floors_total'] = df['floors_total'].fillna(df['floors_total'].median())
```

```python
# Проверяем:

df['floors_total'].isna().sum()
```

С общей этажностью всё, процент пропусков ничтожно мал, замена медианой не исказит данные

#### living_area

```python
# Смотрим пропуски в столбце:

df['living_area'].isna().sum()
```

```python
# К сожалению, в столбце большое количество — 8%. Показатель один из ключевых, отбросить 8% данных нельзя. 
# Посмотрим на состав содержимого:

df['living_area'].describe()
```

Вычислим отношение жилой площади к общей, поместим результаты вычисления в столбец `coeff_living_area`

```python
df['coeff_living_area'] = df['living_area'] / df['total_area']
df['coeff_living_area'].head()
```

```python
# Теперь возьмем медианное значение коэффицента

coeff_living_area = df['coeff_living_area'].median()
coeff_living_area
```

```python
# Заполним пропуски в 'living_area' с помощью полученного коэффицента:

df['living_area'] = df['living_area'].fillna(df['total_area'] * coeff_living_area)

# проверим пропуски еще раз:

df['living_area'].isna().sum()
```

#### is_apartment

```python
# В столбце большое количество пропусков:

df['is_apartment'].isna().sum()
```

```python
# Какие данные есть в заполненных ячейках:

df['is_apartment'].value_counts()
```

```python
# Предположительно при выгрузке данных False означало "не аппартаменты", True — "аппртаменты", а пропущенные 
# значения "не аппартаменты". Пропуски заполним на "0" и приведем весь столбец к типу bool через astype:

df['is_apartment'].fillna(0, inplace=True)
df['is_apartment'] = df['is_apartment'].astype('bool')
df['is_apartment'].value_counts()
```

Теперь всё правильно — 50 True (т.е. это аппартаменты), остальное — False (не аппартаменты)

### Поиск дубликатов

Интересуют неявные дубликаты в поле `locality_name`, они могут помешать нам при определении населенного пункта

```python
# Смотрим значения:

a = df['locality_name'].unique()
sorted(a)
```

```python
# У нас дублируются Мурино, поселок Мурино, посёлок Мурино, Кудрово, деревня Кудрово

df.locality_name = df.locality_name.str.replace('ё', 'е')
df.locality_name = df.locality_name.replace({'Мурино': 'поселок Мурино', 'Кудрово': 'деревня Кудрово'})
a = df['locality_name'].unique()
sorted(a)
```

```python
# Для удобства работы избавимся от "деревень", ПГТ и сел. Напишем функцию:

def locality_name_determine(row):
    lemmas=row['locality_name']
    stop_words=['коттеджный поселок ', 'коттеджный посёлок ', 'деревня ', \
                'поселок городского типа ','посёлок городского типа ',\
                'поселок железнодорожной станции ', 'посёлок железнодорожной станции '
                'поселок станции ', 'посёлок станции '\
                'поселок при железнодорожной станции ', 'посёлок при железнодорожной станции ',\
                'садоводческое некоммерческое товарищество ', 'садовое товарищество ',\
                'городской поселок ', 'городской посёлок ',
                'поселок ','посёлок ', 'село ']
    for word in stop_words:
        if word in lemmas: 
            lemmas=lemmas.replace(word, '')
    return lemmas

# Применим ее к locality_name:

df['locality_name'] = df.apply(locality_name_determine, axis=1)

# Смотрим значения:

a = df['locality_name'].unique()
sorted(a)
```

### Определение категории этажа
